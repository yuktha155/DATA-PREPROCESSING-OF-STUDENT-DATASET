```python
df['Age'] = df['Age'].fillna(df['Age'].mean())
```

```python
missing_values_after_age_imputation = df.isnull().sum()
print('Missing values after Age imputation:')
print(missing_values_after_age_imputation)
```

```python
df['Parental_Education'] = df['Parental_Education'].fillna(df['Parental_Education'].mode()[0])
```

IDENTIFYING WHICH NEEDS TO BE FILLED BY MEAN AND MODE

```python
numerical_cols = df.select_dtypes(include=['number']).columns
categorical_cols = df.select_dtypes(include=['object', 'category']).columns

print("Columns for Mean Filling")
for col in numerical_cols:
    if df[col].isnull().sum() > 0:
        print(col)

print("\nColumns for Mode Filling")
for col in categorical_cols:
    if df[col].isnull().sum() > 0:
        print(col)
```

```python
num_cols = ['Age', 'Hours_Studied', 'Attendance', 'Final_Score']

for col in num_cols:
    mean_value = df[col].mean()
    df[col] = df[col].fillna(mean_value)
```

```python
cat_cols = ['Gender', 'Parental_Education']

for col in cat_cols:
    mode_value = df[col].mode()[0]
    df[col] = df[col].fillna(mode_value)
```
